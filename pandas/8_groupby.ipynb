# Group By

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/bestsellers-with-categories.csv')
```

```python
# agrupa y aplica la operación que le indico columna por columna
# nota: algunas operaciones pueden no tener sentido
df.groupby('Author').count().loc['Walter Isaacson']
```

```python
# Varias funciones en una sola consulta
df.groupby('Author').agg(['min','max'])
```

```python
df.groupby('Author').agg({'Price':['min','max']})
```

```python
df.groupby(['Author','Year']).agg({'Price':['min','max']})
```
