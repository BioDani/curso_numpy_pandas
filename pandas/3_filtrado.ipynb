# Filtrado con iloc y loc

Para navegar en un dataframe

```python
import pandas as pd
```

```python
df = pd.read_csv('./data/bestsellers-with-categories.csv')
```

```python
df[0:4]
```

```python
df['Name'][0:4]
```

```python
df[['Name','Author']]
```

## Loc

Incluye el limite que le defino

```python
# Incluye el index 4
df.loc[0:4]
```

```python
# Pero quiero el nombre y el autor
df.loc[0:4, ['Name','Author']]
```

```python
# True cuando el author sea JJ Smith
df.loc[0:4, ['Author']] == 'JJ Smith'
```

## iloc

Parecido a loc, pero en este caso se accede a travÃ©s del index

```python
# Todas las filas, pero solo las columnas de 0 a 3
df.iloc[:,0:3]
```
